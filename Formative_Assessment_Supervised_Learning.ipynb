# Finding  outliers
##it is necessary for better data and it will avoid unneccasary valuees which is not essesntial but when it comes to medical data alll type of information is neccessary thats why i am not remobing outliers

```python
# iqr method

# Calculate IQR for each feature
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3 - Q1

# Determine outliers
outliers = ((x < (Q1 - 1.5 * IQR)) | (x > (Q3 + 1.5 * IQR)))
outlier_counts = outliers.sum()
print(f'Outlier Counts per Feature:\n{outlier_counts}')
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
# Scale using StandardScaler
standard_scaler = StandardScaler()
X_train_scaled = standard_scaler.fit_transform(x_train)
X_test_scaled = standard_scaler.transform(x_test)
```

# Building model using different alogorthims


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train Multiple Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()

}

# model Evaluvvation

for model_name,model in models.items():
  model.fit(X_train_scaled,y_train)
  y_pred=model.predict(X_test_scaled)
  accuracy=accuracy_score(y_test,y_pred)
  print(f"{model_name} Accuracy: {accuracy:.4f}")  # Print accuracy
  print(classification_report(y_test, y_pred))  # Print classification report
```
